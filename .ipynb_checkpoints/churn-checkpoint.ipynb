**Many of the columns are yes, no etc. Let's print unique values in object columns to see data values**

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

**Some of the columns have no internet service or no phone service, that can be replaced with a simple No**

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

**Convert Yes and No to 1 or 0**

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace = True)
```

```python
for col in df1:
    print(f'{col} : {df1[col].unique()}')
```

```python
df1 = pd.get_dummies(df1, columns=["InternetService", "Contract", "PaymentMethod"])
```

```python
df1.head()
```

```python
df1 = df1.astype({col: 'int8' for col in df1.select_dtypes('bool').columns})
```

```python
df1.sample(5)
```

```python
df1.dtypes
```

```python
col_to_scale = ["tenure","MonthlyCharges","TotalCharges"]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df1[col_to_scale]=scaler.fit_transform(df1[col_to_scale])
```

```python
df1.sample(10)
```

```python
df1["MonthlyCharges"].unique()
```

```python
df1.columns
```

```python
y = df1.Churn
y.sample(5)
```

```python
X = df1.drop(columns = ["Churn"],axis = 1)
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.2,random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,), activation='relu'),  # Hidden layer, not input layer
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs = 100)
```

```python
yp = model.predict(X_test)
yp[0:5]
```

```python
y[0:5]
```

```python
y_pred =[]
for element in yp:
    if element>0.5:
        y_pred.append(1)

    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels = y_test,predictions = y_pred)
plt.figure(figsize=(8, 6))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
